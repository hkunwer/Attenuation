```python
#Linear Regression

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Assuming 'df' is your DataFrame
# Let's map the frequency bands to their numerical midpoints
freq_band_mapping = {
    '(0.1, 0.25)': 0.175,
    '(0.25, 0.5)': 0.375,
    '(0.5, 0.75)': 0.625,
    '(0.75, 1.0)': 0.875,
    '(1.0, 1.25)': 1.125
}
df['frequency_value'] = df['frequency bands'].map(freq_band_mapping)

# Set up the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Assign axes
x = df['distance']
y = df['frequency_value']
z = df['max amplitude (normalized)']

# Plot the actual data points
ax.scatter(x, y, z, color='r', label='Data points')

# Fit the regression model
X = df[['distance', 'frequency_value']]
model = LinearRegression().fit(X, z)

# Predict for the surface
x_surf, y_surf = np.meshgrid(np.linspace(x.min(), x.max(), 100), np.linspace(y.min(), y.max(), 100))
z_surf = model.predict(np.array([x_surf.ravel(), y_surf.ravel()]).T).reshape(x_surf.shape)

# Plot the surface
ax.plot_surface(x_surf, y_surf, z_surf, color='b', alpha=0.5, rstride=100, cstride=100, label='Regression plane')

# Labels and title
ax.set_xlabel('Distance')
ax.set_ylabel('Frequency Band Value')
ax.set_zlabel('Max Amplitude (Normalized)')
plt.title('3D Plot with Regression Plane')

# Display the plot
plt.legend()
plt.show()

# The model's coefficients represent the slopes in the direction of each independent variable
distance_slope = model.coef_[0]
frequency_slope = model.coef_[1]
print(f"Slope in the distance direction: {distance_slope}")
print(f"Slope in the frequency band direction: {frequency_slope}")

# Compute the magnitude of the gradient vector (slope)
overall_slope = np.sqrt(distance_slope**2 + frequency_slope**2)
print(f"Overall slope considering both distance and frequency: {overall_slope}")
```

```python
# Model Selection and evaluation example
# Still need to apply this to mine but this is basic idea if i did it right.
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create synthetic data for demonstration purposes
np.random.seed(42)  # For reproducibility
x = np.random.rand(100, 1) * 10  # Distance
y = np.random.rand(100, 1) * 5   # Frequency
z = 3 * np.log(x) + np.sin(y) + np.random.randn(100, 1)  # Max amplitude (normalized)

# Train a linear model
lin_reg = LinearRegression()
lin_reg.fit(np.hstack((x, y)), z)

# Train a polynomial model
degree = 2
poly_reg = make_pipeline(PolynomialFeatures(degree), LinearRegression())
poly_reg.fit(np.hstack((x, y)), z)

# Train a neural network model
nn_model = Sequential([
    Dense(64, input_dim=2, activation='relu'),
    Dense(64, activation='relu'),
    Dense(1, activation='linear')
])
nn_model.compile(optimizer='adam', loss='mean_squared_error')
nn_model.fit(np.hstack((x, y)), z, epochs=500, verbose=0)

# Create a mesh grid for plotting surfaces
x_surf, y_surf = np.meshgrid(np.linspace(x.min(), x.max(), 50), np.linspace(y.min(), y.max(), 50))
xy_surf = np.hstack((x_surf.reshape(-1, 1), y_surf.reshape(-1, 1)))

# Predict with each model
z_lin_surf = lin_reg.predict(xy_surf).reshape(x_surf.shape)
z_poly_surf = poly_reg.predict(xy_surf).reshape(x_surf.shape)
z_nn_surf = nn_model.predict(xy_surf).reshape(x_surf.shape)

# Plot the observed data and the models' surfaces
fig = plt.figure(figsize=(18, 5))

# Linear model plot
ax1 = fig.add_subplot(131, projection='3d')
ax1.scatter(x, y, z, color='r', label='Data points')
ax1.plot_surface(x_surf, y_surf, z_lin_surf, color='b', alpha=0.3, label='Linear model')
ax1.set_title('Linear Regression')

# Polynomial model plot
ax2 = fig.add_subplot(132, projection='3d')
ax2.scatter(x, y, z, color='r', label='Data points')
ax2.plot_surface(x_surf, y_surf, z_poly_surf, color='g', alpha=0.3, label='Polynomial model')
ax2.set_title('Polynomial Regression')

# Neural network plot
ax3 = fig.add_subplot(133, projection='3d')
ax3.scatter(x, y, z, color='r', label='Data points')
ax3.plot_surface(x_surf, y_surf, z_nn_surf, color='m', alpha=0.3, label='Neural network')
ax3.set_title('Neural Network')

for ax in [ax1, ax2, ax3]:
    ax.set_xlabel('Distance')
    ax.set_ylabel('Frequency')
    ax.set_zlabel('Max Amplitude (Normalized)')

plt.tight_layout()
plt.show()
```
