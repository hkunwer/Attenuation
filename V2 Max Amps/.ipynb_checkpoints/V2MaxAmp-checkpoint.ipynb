```python
from obspy import UTCDateTime
from obspy.taup import TauPyModel
import pandas as pd
from rtergpy.run import defaults, event, etime2name
from rtergpy.waveforms import getwaves
from V2Functions import (
    check_existing_dataframe, processANSStxt, filter_stream_for_velocity_squared,
    maxamp_calc_freq_bands)
# Initialize TauPyModel for seismic wave travel time calculations
model = TauPyModel(model="iasp91")
```

```python
# Load configuration and event defaults
Defaults = defaults()
Event = event()
Defaults.src='IRIS'
Defaults.network='??'
Defaults.chan= 'BHZ'
Defaults.stationrange=[1,10.0]
Event.ecount='00'
# Defaults.src = 'RASPISHAKE'
# Defaults.network = 'AM'
# Defaults.chan = 'EHZ'
# Defaults.stationrange = [0.5, 10.0]
# Event.ecount = '00'
Event.newData = False  # False means Use already downloaded data, True new data
```

```python
# Process or load ANSS event data
#ANSS = processANSStxt() #If you have raw ANSS list, process with this
ANSS = pd.read_csv('ANSS_processed_data.csv')  # comment out top if already processed to load with this
```

```python
normalized_df = check_existing_dataframe()  # Check for existing DataFrame to append results
```

```python
edateold = ""

for index, EQ in ANSS.iterrows():
    eloc = [EQ.Latitude, EQ.Longitude, EQ.Depth]
    magnitude = {'type': EQ.Mtype, 'value': EQ.Mag}
    year, mo, dy = EQ.Date.split('-')
    hh, mn, sec = EQ.Time.split(':')
    etime = UTCDateTime(int(year), int(mo), int(dy), int(hh), int(mn), float(sec))

    Event.ecount = str(int(Event.ecount) + 1).zfill(2) if EQ.Date == edateold else '00'
    Event.iter = 'RS' if Defaults.src == 'RASPISHAKE' else 'IRIS'
    edateold = EQ.Date
    Event.eventname = etime2name(etime, ecount=Event.ecount) + Event.iter
    Event.origin = [eloc, etime]

    print(f"\n\n{Event.eventname} ===============================")

    try:
        try:
            st, df = getwaves(Defaults, Event)
            print("Completed getwaves")
        
        except Exception as e:
            print(f"ERROR: {Event.eventname} could not complete getwaves due to {e}")
            continue
        
        # Check if the stream has at least 5 traces (This ensures a more reliable normalization)
        if len(st) < 5:
            print("Stream contains fewer than 5 traces. Process Cancelled.")
            continue  # Skip the rest of the current loop iteration and move to the next

        stp = filter_stream_for_velocity_squared(st)        
        maxamp_calc_freq_bands(stp, EQ, Defaults, etime, eloc)
        
    except Exception as e:
        print(f"ERROR: {Event.eventname} could not finish processing due to {e}")
```

```python
import numpy as np
import pickle
import pandas as pd
import matplotlib.pyplot as plt
freqlist = ['(0.1, 0.25)', '(0.25, 0.5)', '(0.5, 0.75)', '(0.75, 1.0)', '(1.0, 1.25)']
colors = ['b', 'g', 'r', 'c', 'm']  # Colors for the outlines of the plots

plt.figure(figsize=(10, 6))  # Set the figure size

# Clean the DataFrame to ensure the frequency band column is a string
df['frequency band'] = df['frequency band'].astype(str)

for freq, color in zip(freqlist, colors):
    # Since your frequency bands are stored as strings, we'll need to ensure we're comparing strings
    # Ensure this representation matches the one in your DataFrame
    freq_band_df = df[df['frequency band'] == freq]

    if not freq_band_df.empty:
        # Plot max amplitude against distance for the filtered DataFrame with empty fill
        plt.scatter(freq_band_df['distance'], freq_band_df['max amplitude'],
                    edgecolors=color, facecolors='none', label=f"Freq band {freq}")
    else:
        print(f"No data for frequency band {freq}")

plt.title('Max Amplitude vs. Distance for Different Frequency Bands')
plt.xlabel('Distance (km)')
plt.ylabel('Max Amplitude')
plt.legend()
plt.show()
```
