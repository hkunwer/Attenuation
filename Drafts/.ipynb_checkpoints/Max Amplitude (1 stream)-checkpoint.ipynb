```python
trace = stp[0]
data = trace.data                           # the actual data
samp = data.shape[-1]                       # number of sample point
delta = trace.stats.delta                  # sample spacing 
time = trace.times()                       # time axis 

# ---- compute fft for our data
fftdata = np.fft.rfft(data, n=samp)

# ---- Fdat has real and complex part, to compute power spectrum we take the
fftdata_abs = np.abs(fftdata)

# ---- compute x-axis in frequency-domain for plotting
xf = np.linspace(0.0, 1.0/(2.0*delta), int(samp/2)+1)

# ---- get the max value of amplitude spectrum
idx_of_max_amp = np.where(fftdata_abs == fftdata_abs.max()) # find the index of max amplitudes
```

```python
# ---- plot the waveform
plt.subplot(211)
plt.title('Time Domain')
plt.plot(time, data, label="Original Data", color='b', linewidth=1.5)
plt.legend(loc='upper center')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
```

```python
# ---- ploting frequency against amplitude
plt.subplot(212)
plt.title('Frequency Domain')
plt.plot(xf, 1.0/samp * fftdata_abs, color='b', linewidth=1.5) # NOTICE the nor
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude')
plt.yscale('log')
plt.xscale('log')
plt.tight_layout()
frq_max_amp = xf[idx_of_max_amp]
amp_max_amp = 1.0/samp * fftdata_abs[idx_of_max_amp]
plt.scatter(frq_max_amp, amp_max_amp, color='red')
plt.show()
```
