```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from datetime import datetime

# Create a folder to save plots if it doesn't exist
save_folder_base = 'plot'
if not os.path.exists(save_folder_base):
    os.makedirs(save_folder_base)

pd.set_option('display.float_format', lambda x: '%.15f' % x)

# Assuming df is your DataFrame

# Convert 'event' column to string type
df.loc[:, 'event'] = df['event'].astype(str)

# Get unique events (dates) from the filtered data
unique_events = df['event'].unique()

# Filter only the frequency band (1.0, 1.25) as a reference
normalizeBench = 200
filtered_dataframe = df[df['frequency band'] == (1.0, 1.25)]

# Calculate the best fit line for the log-transformed filtered data
filtered_log_max_amplitude = np.log10(filtered_dataframe['max amplitude'])
filtered_coefficients = np.polyfit(filtered_dataframe['distance'], filtered_log_max_amplitude, 1)
filtered_best_fit_line = np.poly1d(filtered_coefficients)

# normalization factor
normalization_factor = filtered_best_fit_line(normalizeBench)

for target_event in unique_events:
    
    # Convert 'frequency band' to a tuple with the correct format
    filtered_df = df[df['frequency band'] == (1.0, 1.25)]

    # Filter for the target event in the specified frequency band
    filtered_df = filtered_df[filtered_df['event'] == target_event]

    
    if filtered_df.empty:
        continue

    fig, axs = plt.subplots(figsize=(8, 6))

    for event in [target_event]:
        event_data = filtered_df[filtered_df['event'] == event]

        axs.scatter(
            event_data['distance'],
            np.log10(event_data['max amplitude']),
            label=event)
        axs.set_xlabel('Distance')
        axs.set_ylabel('Max amplitude (log scale)')
        axs.set_title(f'Max amplitude for Frequency Band (1.0, 1.25)')
        axs.grid(True)
        # Display the legend
        axs.legend()

        # Apply best fit line only for the 1.0-1.25Hz frequency band
        if not event_data.empty:
            # Calculate the best fit line for the current event
            x_values = event_data['distance']
            y_values = np.log10(event_data['max amplitude'])
            coefficients = np.polyfit(x_values, y_values, 1)
            best_fit_line = np.poly1d(coefficients)

            # Add best fit line to the scatter plot with a star marker for the normalization benchmark
            axs.plot(x_values, best_fit_line(x_values), color='red', label=None)
            axs.scatter(normalizeBench, best_fit_line(normalizeBench), marker='*', s=200, color='red', label=f'Normalization Benchmark')

            # Calculate normalization factor for the current event
            normalization_factor_event = best_fit_line(normalizeBench)

            # Print normalization factor and best fit line equation in the bottom left corner of the plot
            axs.text(0.05, 0.05, f'Normalization Factor: {normalization_factor_event:.4f}\nBest Fit Line Equation: y = {coefficients[0]:.4f}x + {coefficients[1]:.4f}', transform=axs.transAxes, fontsize=10, verticalalignment='bottom', color='red')
        
        # Extract the year from the event timestamp
        event_timestamp = pd.to_datetime(event)
        year_folder = str(event_timestamp.year)

        # Create a subfolder for the year if it doesn't exist
        save_folder_year = os.path.join(save_folder_base, year_folder)
        if not os.path.exists(save_folder_year):
            os.makedirs(save_folder_year)

        # Create a subfolder for the event within the year folder
        save_folder_event = os.path.join(save_folder_year, event)
        if not os.path.exists(save_folder_event):
            os.makedirs(save_folder_event)

        # Save the plot if there are more than 5 traces (event occurrences)
        if len(event_data) > 5:
            # Construct the save path with the event name in the event subfolder
            save_path = os.path.join(save_folder_event, f'Normalization of {event}.png')

            # Save the plot
            plt.savefig(save_path, bbox_inches='tight')

        # Display the plot (comment this line if you don't want to display the plot)
        plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

def max_and_normalized_max_plots(df):
    freqlist = [(0.10, 0.25), (0.25, 0.50), (0.50, 0.75), (0.75, 1.00), (1.00, 1.25)]

    # Create a base folder to save plots
    base_folder = 'plot'
    all_events_folder = 'All_Events'
    if not os.path.exists(base_folder):
        os.makedirs(base_folder)

    # Create a subfolder for 'All_Events'
    save_folder_all_events = os.path.join(base_folder, all_events_folder)
    if not os.path.exists(save_folder_all_events):
        os.makedirs(save_folder_all_events)

    for freq_band in freqlist:
        filtered_df = df[df['frequency band'] == freq_band]
        filtered_df.loc[:, 'event'] = filtered_df['event'].astype(str)

        # Get unique events (dates) from the filtered data
        unique_events = filtered_df['event'].unique()

        # Create a figure with two subplots
        fig, axs = plt.subplots(1, 2, figsize=(12, 5))

        for event in unique_events:
            event_data = filtered_df[filtered_df['event'] == event]

            # Only plot events with more than 3 traces
            if len(event_data) > 3:
                axs[0].scatter(
                    event_data['distance'],
                    np.log10(event_data['max amplitude']),
                    label=event)
                axs[1].scatter(
                    event_data['distance'],
                    event_data['max amplitude (normalized)'],
                    label=event)

        axs[0].set_xlabel('Distance')
        axs[0].set_ylabel('Max amplitude')
        axs[0].set_title(f'Max amplitude for Frequency Band {freq_band}')
        axs[0].legend()
        axs[0].grid(True)

        axs[1].set_xlabel('Distance')
        axs[1].set_ylabel('Max amplitude (normalized)')
        axs[1].set_title(f'Max amplitude (normalized) for Frequency Band {freq_band}')
        axs[1].legend()
        axs[1].grid(True)

        plt.tight_layout()

        # Create a subfolder for the frequency band within 'All_Events'
        save_folder_band = os.path.join(save_folder_all_events, f'{freq_band[0]}_{freq_band[1]}')
        if not os.path.exists(save_folder_band):
            os.makedirs(save_folder_band)

        # Save the plot
        save_path = os.path.join(save_folder_band, f'Max_Normalized_Plots_{freq_band[0]}_{freq_band[1]}.png')
        plt.savefig(save_path, bbox_inches='tight')

        # Close the figure to avoid displaying the plot
        plt.close()
```
